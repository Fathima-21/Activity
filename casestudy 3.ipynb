# 2.build a regression model to predict profit from given data and comment your findings on it

```python
y=df['Profit(in Thousands)']
x=df.drop(['Profit(in Thousands)','Date'],axis=1)
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=75,test_size=0.25)
```

```python
from sklearn import linear_model
lr=linear_model.LinearRegression()
model=lr.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
print(y_test)
print(predictions)
```

```python
from sklearn.metrics import mean_squared_error
print('MSE is',mean_squared_error(y_test,predictions))
```
