# 3.who scores the most on average for math,reading and writing based on 
   a. gender
   b. test preparation course

```python
score1=df.groupby('gender')['math score'].mean().nlargest(1)
score2=df.groupby('gender')['reading score'].mean().nlargest(1)
score3=df.groupby('gender')['writing score'].mean().nlargest(1)
score1
score2
score3
```

```python
score1=df.groupby('test preparation course')['math score'].mean().nlargest(1)
score2=df.groupby('test preparation course')['reading score'].mean().nlargest(1)
score3=df.groupby('test preparation course')['writing score'].mean().nlargest(1)
score1
score2
score3
```

# 4.what do you think about scoring variation for math,reading and writing based on
    a.gender
    b.test preparation course

```python
score1=df.groupby('gender')['math score','reading score','writing score'].var()
score1
```

# for female and male ,scoring variation is high for math score and least for reading score

```python
score1=df.groupby('test preparation course')['math score','reading score','writing score'].var()
score1
```

# for those who completed course,scoring variation is high in math score and least in writing score.for those who didnt complete,it is high in math score and least in reading score
